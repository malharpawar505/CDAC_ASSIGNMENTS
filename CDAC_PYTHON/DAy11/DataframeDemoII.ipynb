```python
d={'id':[1,2,3],'name':['kirti','rashmi','rakesh']}
d1={'id':[1,2,4],'age':[12,34,57]}
df=pd.DataFrame(d)
df1=pd.DataFrame(d1)
print(df)
print(df1)
```

```python
df3=pd.merge(df,df1,on="id",how="inner")
df3
```

```python
df3=pd.merge(df,df1,on="id",how="left")
df3
```

```python
df3=pd.merge(df,df1,on="id",how="right")
df3
```

```python
df3=pd.merge(df,df1,on="id",how="outer")
df3
```
